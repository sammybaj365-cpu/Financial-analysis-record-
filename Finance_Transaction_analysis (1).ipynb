```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker 
money_fmt = mticker.FuncFormatter(lambda x, _: f"${x:,.0f}")
print('Ready!')
```

```python
df = pd.read_excel('finance_transactions_dataset.xlsx')

print(df.shape)       
print(df.columns.tolist())
```

```python
df.dropna(subset=['Amount', 'Net_Amount', 'Transaction_Date', 
'Approval_Status'],
          inplace=True)
print('Rows remaining:', len(df))
```

```python

df['Company']    = df['Company'].fillna('Unknown')
df['Department'] = df['Department'].fillna('Unknown')
df['Analyst']    = df['Analyst'].fillna('Unknown')
df['Category']   = df['Category'].fillna('N/A')
print('missing value handled')
```

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'], 
errors='coerce')
df.dropna(subset=['Transaction_Date'], inplace=True)
df['Year']  = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.to_period('M').astype(str)
print('Sample dates:')
print(df[['Transaction_Date', 'Year', 'Month']].head(3))
```

```python
df.drop_duplicates(inplace=True)
print('Final row count:', len(df))
```

```python
# which transaction moves more money

df.groupby('Transaction_Type')['Amount'].sum().sort_values().plot(kind='bar', 
color='green')
plt.title('Total Amount by Transaction Type')
plt.ylabel('Total Amount ($)')
plt.xticks(rotation=15)
plt.gca().yaxis.set_major_formatter(money_fmt)  
plt.tight_layout()
plt.show()
```

```python
# Net Amount vs Tax Amount by Sector  (Grouped Bar)
# How much tax does each sector pay vs what it keeps?

df.groupby('Sector')[['Net_Amount', 'Tax_Amount']].sum().plot(kind='bar', 
color=['steelblue', 'coral'])
plt.title('Net Amount vs Tax Amount by Sector')
plt.ylabel('Amount ($)')
plt.xticks(rotation=20)
plt.gca().yaxis.set_major_formatter(money_fmt)   
plt.tight_layout()
plt.show()
```

```python
# Monthly Transaction Volume  (Line)
# Is financial activity growing or shrinking over time?

df.groupby('Month')['Amount'].sum().plot(kind='line', marker='o', 
color='coral')
plt.title('Monthly Transaction Volume')
plt.ylabel('Total Amount ($)')
plt.xticks(rotation=45)
plt.gca().yaxis.set_major_formatter(money_fmt)   
plt.tight_layout()
plt.show()
```

```python
# Approval Status Distribution  (Pie)
# What share of transactions are approved, pending, or rejected?

df['Approval_Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', 
startangle=140)
plt.title('Approval Status Distribution')
plt.ylabel('')   
plt.tight_layout()
plt.show()
```

```python
# Top 5 Analysts by Amount Processed  (Horizontal Bar)
# Which analyst handled the largest transaction volume?

df.groupby('Analyst')['Amount'].sum().nlargest(5).plot(kind='barh', 
color='mediumseagreen')
plt.title('Top 5 Analysts by Total Amount')
plt.xlabel('Total Amount ($)')
plt.gca().xaxis.set_major_formatter(money_fmt)   
plt.tight_layout()
plt.show()
```
